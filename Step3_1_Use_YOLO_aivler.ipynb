```python
os.makedirs(f'{data_path}/images', exist_ok=True)
os.makedirs(f'{data_path}/labels', exist_ok=True)
```

```python
import glob
from tqdm import tqdm
```

```python
db_list = ['Face_Recognition_1', 'Face_Recognition_2']

for db in db_list:
  rm_tree = glob.glob(f'{data_path}/{db}/*/labels_edit')
  for i, _ in enumerate(rm_tree):
    shutil.rmtree(rm_tree[i])
```

```python
db_list = ['Face_Recognition_1', 'Face_Recognition_2']
for db in db_list:
    print(f'Modifying labels in {db}')
    labels_fp_list = glob.glob(f'{data_path}/{db}/*/labels/*')

    for labels_fp in tqdm(labels_fp_list):
        new_lines = []
        with open(labels_fp, 'r') as f:
            lines = f.readlines()
            for i, line in enumerate(lines):
                elements = line.split(' ')
                elements[0] = '1'
                new_lines.append( ' '.join(elements) )

        labels_fp_split = labels_fp.split('/')
        os.makedirs(f'{data_path}/{db}/{labels_fp_split[-3]}/labels_edit', exist_ok=True)
        with open(f'{data_path}/{db}/{labels_fp_split[-3]}/labels_edit/{labels_fp_split[-1]}', 'w') as f:
            f.write(''.join(new_lines))
```

#### 3) 각 폴더에 데이터 옮기기

```python
import random
```

```python
# db_list = ['Face_Recognition_1', 'Face_Recognition_2']
db_list = ['Face_Recognition_2']
val_ratio = 0.2
random.seed(2024)

for db in db_list:
  fp_list_w_jpg = glob.glob(f'{data_path}/{db}/*/images/*.jpg')
  random.shuffle(fp_list_w_jpg)
  fp_list_w_txt = ['/'.join(jpg_fp.split('/')[:-2]) + '/labels/' + os.path.splitext(os.path.basename(jpg_fp))[0] + '.txt' for jpg_fp in fp_list_w_jpg]

  val_num = int(len(fp_list_w_jpg) * val_ratio)
  train_jpg_list = fp_list_w_jpg[val_num:]
  train_txt_list = fp_list_w_txt[val_num:]
  val_jpg_list = fp_list_w_jpg[:val_num]
  val_txt_list = fp_list_w_txt[:val_num]

  print(f'# trainset from {db} :', len(train_jpg_list))
  print(f'# valset from {db} :', len(val_jpg_list))

  print(f'Copying trainset from {db}...')
  for i, train_jpg in tqdm(enumerate(train_jpg_list)):
    filename = os.path.splitext(os.path.basename(train_jpg))[0]
    if (filename == os.path.splitext(os.path.basename(train_txt_list[i]))[0]):
      shutil.copy(src=train_jpg, dst=f'{data_path}/images/train/{filename}.jpg')
      shutil.copy(src=train_txt_list[i], dst=f'{data_path}/labels/train/{filename}.txt')
  print('Done!')

  print(f'Copying valset from {db}...')
  for i, val_jpg in tqdm(enumerate(val_jpg_list)):
    filename = os.path.splitext(os.path.basename(val_jpg))[0]
    if (filename == os.path.splitext(os.path.basename(val_txt_list[i]))[0]):
      shutil.copy(src=val_jpg, dst=f'{data_path}/images/val/{filename}.jpg')
      shutil.copy(src=val_txt_list[i], dst=f'{data_path}/labels/val/{filename}.txt')
  print('Done!')
```
